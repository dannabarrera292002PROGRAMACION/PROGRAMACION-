#EL VALOR PRESENTE NETO


El **Valor Presente Neto (VPN)**, también conocido como *Valor Actual Neto (VAN)*, es una herramienta financiera fundamental para evaluar la rentabilidad de una inversión o proyecto.
***Su función principal es traer al presente todos los flujos de dinero futuros que se esperan de un negocio, restándoles la inversión inicial***. Esto permite saber si, al valor de hoy, el proyecto realmente genera riqueza adicional o si es mejor buscar otra alternativa.

##¿Cómo se interpreta el resultado?



- **VPN > 0 (Positivo):** El proyecto es rentable. Los beneficios superan la inversión inicial y la rentabilidad exigida, por lo que se debe aceptar.
- **VPN = 0 (Cero):** El proyecto no genera ni ganancias ni pérdidas netas; simplemente recupera la inversión y rinde exactamente la tasa de interés aplicada.
- **VPN < 0 (Negativo):** El proyecto destruye valor. La inversión no logra cubrir los costos ni la rentabilidad mínima, por lo que debe rechazarse.  Elemento de la lista





$$VPN = -I_0 + \sum_{t=1}^{n} \frac{F_t}{(1 + k)^t}$$
